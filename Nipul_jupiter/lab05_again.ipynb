```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Bank1 = pd.read_csv('bank-data.csv')
```

```python
Bank1.dtypes
```

```python
Bank1.columns
```

```python
columns=['id']

Bank1.drop(columns,inplace=True,axis=1)
```

```python
Bank1.columns
```

```python
Bank1.head()
```

```python
Bank1.rename(columns={'car' : 'carinfo'},inplace=True)
```

```python
#Construct the plot for pep
sns.countplot(x='pep',data=Bank1)
```

```python
Bank1['pep'].value_counts().plot.bar()
```

```python
#Group the data sex and pep according to the coun

Bank1.groupby(['sex','pep'])['pep'].count()
```

```python

#Construct the graph for pep vs sex

sns.countplot(x='pep',hue='sex',data=Bank1)
```

```python
#Find the count of the attribute children according to the category

Bank1.groupby(['children'])['children'].count()
```

```python
#Construct the graph for attribute children according to the category

sns.countplot(x='children',data=Bank1)
```

```python
#Construct the chart to illustrate pep, children and sex of the customer based on the count

sns.catplot(x='pep',hue='children',col='sex',data=Bank1,kind='count')
```
